## 4. Säulendiagramm mit Altersverteilung
Stellen Sie die Altersverteilung der Beschäftigten in einem gruppierten Säulendiagramm dar, in dem Männer und Frauen getrennt ausgewiesen werden.<br>
Es sollen die Altersklassen bis (einschließlich) 20 Jahre, 21 – 30 Jahre, 31 – 40 Jahre, 41 – 50 Jahre und über 50 Jahre gewählt werden.

***Tipp:*** Sie benötigen hierfür die Python Bibliothek **Matplotlib** und **NumPy** sowie aus **Pandas** die Funktion `value_counts()`. <br>
***Beispiellösung:***
<div>
<img src="https://drive.google.com/uc?id=1f7Vz3JuSa958w8-dBTnPSI24hJxWa17F" width="300"/>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

altersgrenzen=[0,20,30,40,50,200]
width=0.35



altersklassen=['(0,20]','(21,30]','(31,40]','(41,50]','(50,200]']
data['altersgrenzen']= pd.cut(data['Alter'], bins=altersgrenzen, labels= altersklassen)



data_hf = pd.value_counts(data['Alter'], sort=False).rename_axis('Alter').reset_index(name='Häufigkeit')
frauen_data = data.loc[data['Anrede']=='Frau', ['Alter', 'altersgrenzen']]
frauen_data = frauen_data.reset_index()
frauen_data = frauen_data.drop('index', axis=1)

maenner_data = data.loc[data['Anrede']== 'Herr', ['Alter', 'altersgrenzen']]
maenner_data = maenner_data.reset_index()
maenner_data = maenner_data.drop('index', axis=1)

frauen_hf = pd.value_counts(frauen_data['altersgrenzen'], sort=False).rename_axis('Alter').reset_index(name='Häufigkeit')
maenner_hf = pd.value_counts(maenner_data['altersgrenzen'], sort=False).rename_axis('Alter').reset_index(name='Häufigkeit')


plt.bar(np.arange(len(frauen_hf['Alter'])), frauen_hf['Häufigkeit'], width, color='darkblue', label='Frauen')
plt.bar(np.arange(len(maenner_hf['Alter'])) + width, maenner_hf['Häufigkeit'], width, color='orange', label='Männer')




plt.xlabel('Alter')
plt.ylabel('Häufigkeit')
plt.title('Altersverteilung Männer und Frauen zum Stichtag 31.12.2005')
plt.xticks(np.arange(len(frauen_hf['Alter'])) + width / 2, altersklassen)
plt.legend()


               


plt.show()   
```

## 6. Kennzahlen
Berechnen Sie **Mittelwert**, **empirische Varianz**, **empirische Standardabweichung**, **Median** und **Spannweite** der Alterszahlen: 
- für alle Beschäftigen und
- getrennt für Männer und Frauen.

***Beispiellösung:***
<div>
<img src="https://drive.google.com/uc?id=1adhKBVlds5OlO-9B5CClT2ybPolv5wVW" width="300"/>
</div>

```python


if "Anrede" in data.columns:
    frauen_data = data[data['Anrede'] == 'Frau']
    maenner_data = data[data['Anrede'] == 'Herr']
else:
    frauen_data = data[data['Geschlecht'] == 'weiblich']
    maenner_data = data[data['Geschlecht'] == 'männlich']

mittelwert = str(round(data['Alter'].mean(), 4))
mittelwert_frauen = str(round(frauen_data['Alter'].mean(), 4))
mittelwert_herren = str(round(maenner_data['Alter'].mean(), 4))

varianz = str(round(data['Alter'].var(), 4))
varianz_frauen = str(round(frauen_data['Alter'].var(), 4))
varianz_herren = str(round(maenner_data['Alter'].var(), 4))

std = str(round(data['Alter'].std(), 4))
std_frauen = str(round(frauen_data['Alter'].std(), 4))
std_herren = str(round(maenner_data['Alter'].std(), 4))

median = str(round(data['Alter'].median(), 4))
median_frauen = str(round(frauen_data['Alter'].median(), 4))
median_herren = str(round(maenner_data['Alter'].median(), 4))

spannweite = data['Alter'].max() - data['Alter'].min()
spannweite_frauen = frauen_data['Alter'].max() - frauen_data['Alter'].min()
spannweite_herren = maenner_data['Alter'].max() - maenner_data['Alter'].min()

kz = np.array([[mittelwert, mittelwert_frauen, mittelwert_herren],
               [varianz, varianz_frauen, varianz_herren],
               [std, std_frauen, std_herren],
               [median, median_frauen, median_herren],
               [spannweite, spannweite_frauen, spannweite_herren]])

kennzahlen = pd.DataFrame(data=kz, index=['Mittelwert', 'Varianz', 'Std.-Abw.', 'Median', 'Spannweite'],
                          columns=['gesamt', 'Frauen', 'Herren'])

kennzahlen
```

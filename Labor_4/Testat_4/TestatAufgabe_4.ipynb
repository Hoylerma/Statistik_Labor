```python
import pandas as pd

data = pd.read_csv('sr_aufg_4_82_MR1.txt', delim_whitespace=True, decimal=',', header=None, names=['MR1'])
data['MR2'] = pd.read_csv('sr_aufg_4_82_MR2.txt', delim_whitespace=True, decimal=',', header=None)
data['MR3'] = pd.read_csv('sr_aufg_4_82_MR3.txt', delim_whitespace=True, decimal=',', header=None)
data.head(20)
```

## 2. Kennzahlen bestimmen
Berechnen Sie für jeden der drei Messreihen **Mittelwert**, **empirische Varianz**, **empirische Standardabweichung**, **Median** und **Spannweite**.

***Beispiellösung:***<br>
Bitte geben Sie die Kennzahlen in tabellarischer Form ab.

```python
import numpy as np


mittelwert_mr1 = str(round(data['MR1'].mean(), 4))
mittelwert_mr2 = str(round(data['MR2'].mean(), 4))
mittelwert_mr3 = str(round(data['MR3'].mean(), 4))

varianz_mr1 = str(round(data['MR1'].var(),4))
varianz_mr2 = str(round(data['MR2'].var(),4))
varianz_mr3 = str(round(data['MR3'].var(),4))

median_mr1 = str(round(data['MR1'].median(), 4))
median_mr2 = str(round(data['MR2'].median(), 4))
median_mr3= str(round(data['MR3'].median(), 4))

spannweite_mr1 = str(round(data['MR1'].max() - data['MR1'].min(), 4))
spannweite_mr2 = str(round(data['MR2'].max() - data['MR2'].min(),4 ))
spannweite_mr3 = str(round(data['MR3'].max() - data['MR3'].min(), 4))

std_mr1 = str(round(data['MR1'].std(), 4))
std_mr2 = str(round(data['MR2'].std(), 4))
std_mr3 = str(round(data['MR3'].std(), 4))

kz = np.array([[mittelwert_mr1, mittelwert_mr2, mittelwert_mr3],
               [varianz_mr1, varianz_mr2, varianz_mr3],
               [std_mr1, std_mr2, std_mr3],
               [median_mr1, median_mr2, median_mr3],
               [spannweite_mr1, spannweite_mr2, spannweite_mr3]])

kennzahlen = pd.DataFrame(data=kz, index=['Mittelwert', 'Varianz', 'Std.-Abw.', 'Median', 'Spannweite'],
                          columns=['MR1', 'MR2', 'MR3'])

kennzahlen
```

- Welche Normalverteilungsdichte haben Sie gezeichnet, d. h., welche Werte haben Sie für die Parameter $\mu$ bzw. $\sigma$ eingesetzt?

```python
from scipy.stats import norm

fig, axs = plt.subplots(1,1)
mu = float(mittelwert_mr3)
sigma = float(std_mr3)



x_values = np.arange(346, 358, 0.05)
y_values = norm.pdf(x_values, mu, sigma)


max_hist_value = max(np.histogram(werte_mr3, bins=10)[0])
y_values_scaled = y_values * max_hist_value / max(y_values)

#value = max_hist_value / max(y_values)

axs.hist(werte_mr3, label="Histogramm MR3") 
axs.plot(x_values, y_values_scaled, color='r', label="Dichtekurve MR3")  # Plot der Verteilungsfunktion
axs.set_title("Histogramm und Verteilungsfunktion MR3")
axs.set_xlabel("Wert")
axs.set_ylabel("Häufigkeit / Dichte f(x)")
axs.legend()  

plt.show()

```
